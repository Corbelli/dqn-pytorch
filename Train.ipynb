# Training


```python
%matplotlib inline
from unityagents import UnityEnvironment
```

```python
env = UnityEnvironment(file_name="Environments/Banana.app")
# get the default brain
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
```

```python
from dqn import PriorAgent, PTraining

agent = PriorAgent(state_size=37, action_size=4, seed=0)
training_setup = PTraining(n_episodes=1000, eps_start=1, eps_end=0.01, eps_decay=0.995)
scores = training_setup.train(agent, env, brain_name, track_every=2, plot=True, weights='priordqn.pth')
```

```python
from dqn import  Benchmarks

benchs = Benchmarks()
bench_dict = benchs.load_benchmarks()
benchs.plot_benchs(bench_dict, title='DDDQN vs DDQN vs DQN Training', mean=100, opacity=0.1)
```

```python
from dqn import DDQNAgent, Training, Benchmarks

agent = DDQNAgent(state_size=37, action_size=4, seed=0)
agent.load_weights('ddqn.pth')
scores = Training().play(agent, env, brain_name)
```
