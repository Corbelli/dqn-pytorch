# Training


```python
%matplotlib inline
from unityagents import UnityEnvironment
import numpy as np
```

```python
env = UnityEnvironment(file_name="Environments/Banana.app")
# get the default brain
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
```

```python
from dqn import DQNAgent, Training, Benchmarks

agent = DQNAgent(state_size=37, action_size=4, seed=0)
training_setup = Training(n_episodes=1000, eps_start=1, eps_end=0.01, eps_decay=0.995)
scores = training_setup.train(agent, env, brain_name, track_every=2, plot=True)
```

```python
from dqn import Benchmarks

benchs = Benchmarks()
```
